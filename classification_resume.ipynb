```python
import pdfplumber

def pdf_to_string(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Example usage
pdf_path_adarsh = 'Adarsh Mishra-resume.pdf'
pdf_adarsh = pdf_to_string(pdf_path_adarsh)
pdf_path_aryan = 'Aryan Kaushik Resume.pdf'
pdf_aryan= pdf_to_string(pdf_path_aryan)
pdf_path_abhi='Abhishek Verma (1).pdf'
pdf_abhi=pdf_to_string(pdf_path_abhi)
pdf_path_alok='resume_ARD (1).pdf'
pdf_alok=pdf_to_string(pdf_path_alok)
pdf_path_aditya='aditya_3.1.pdf'
pdf_aditya=pdf_to_string(pdf_path_aditya)
pdf_path_ayush='ayush.pdf'
pdf_ayush=pdf_to_string(pdf_path_ayush)
print(pdf_aditya)
```

```python
file=pd.read_csv('resu.csv',nrows=891)
```

```python
classy=file[['Category']]
classx=file[['Resume']]
print(classx)
print(classy)
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
ps = WordNetLemmatizer()
c = []
for i in range(0, len(classx)):
    r=re.sub('[^a-zA-Z]', ' ',classx['Resume'][i])
    r=r.lower()
    r=r.split()

    r=[ps.lemmatize(word) for word in r if not word in stopwords.words('english')]
    r=' '.join(r)
    c.append(r)

c
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
classx = cv.fit_transform(c).toarray()
print(classx)
type(classx)
print(classy)
type(classy)
```

```python
from sklearn.model_selection import train_test_split
xclass_t, xclass_test, yclass_t, yclass_test = train_test_split(classx, classy,test_size = 0.3, random_state =1)
xclass_train, xclass_val, yclass_train, yclass_val = train_test_split(xclass_t, yclass_t, test_size = 0.2,random_state = 2)
xclass_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
f= MultinomialNB().fit(xclass_train, yclass_train)
```

```python
q = pd.read_csv("resu.csv", nrows=1100)
q = q.iloc[792:861]
classX = q['Resume'].values
classY = q['Category'].astype(int).values

r = []
for i in range(len(classX)):
    p = re.sub('[^a-zA-Z]', ' ', classX[i])
    p = p.lower()
    p = p.split()
    p = [ps.lemmatize(word) for word in p if not word in stopwords.words('english')]
    p = ' '.join(p)
    r.append(p)

z = cv.transform(r).toarray()
print(classY)
print(classX)
y_pred_new = f.predict(z)
print(y_pred_new)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(classY, y_pred_new)
print(f'Accuracy: {accuracy}')
```

```python
%%writefile app.py

import streamlit as st
import pickle
import numpy as np

# Load your trained model
with open('model.pkl', 'rb') as file:
    model = pickle.load(file)

# Define a function to make predictions
def predict_category(text):
    prediction = model.predict([text])
    return prediction[0]

# Streamlit app interface
st.title("Resume Category Predictor")

# Input text box
text_input = st.text_area("Enter the resume text here:")

# Predict button
if st.button("Predict"):
    if text_input:
        category = predict_category(text_input)
        st.write(f"The predicted category is: **{category}**")
    else:
        st.write("Please enter some text.")
```
